```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fraud_data.csv')
df = df[['amt', 'category', 'is_fraud']]
total_transaction = len(df['amt'])
unique = {'0','1'}
df['is_fraud'] = df['is_fraud'].apply(lambda x: '0' if x not in unique else x)
fraud_transaction = df['is_fraud'].astype('int')
```

```python
no_of_fraud_transactions = fraud_transaction.sum()
non_fraud_transactions = total_transaction - no_of_fraud_transactions
fraud_percentage = (no_of_fraud_transactions / total_transaction)*100
print(fraud_percentage)
non_fraud_percentage = (non_fraud_transactions / total_transaction)*100
print(non_fraud_percentage)
```

```python
labels = ['Non fraud','Fraud']
sizes = [non_fraud_transactions,no_of_fraud_transactions]
colors = ['#003f5c','#ff6361']
explode = (0, 0)
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
plt.title('Percentage of Fraud vs. Non-Fraud Transactions')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='amt',hue='is_fraud',element='step',stat='density',common_norm=False)
plt.title("Transaction Amounts by fraud status")
plt.xlabel("Amount")
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='category',hue='is_fraud')
plt.title('Transaction categories by Fraud status')
plt.xlabel("Category")
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
